```python
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
import csv
import requests
import pandas as pd
```

```python
url = ("https://wuzzuf.net/search/jobs/?q=data+science&a=hpb")
```

```python
client = urlopen(url)
```

```python
html = client.read()
```

```python
html
```

```python
soup = bs(html, "html.parser")
```

```python
soup
```

```python
containers = soup.find_all("div", {"class": "css-1gatmva e1v1l3u10"})
```

```python
len(containers)
```

```python
bs.prettify(containers[0])
```

```python
containers[0] .div.h2.text
```

```python
file1 = open("E:\python\webscrapping.test.csv", "w")
headers = "job_title , company_name , job_type\n"
file1.write(headers)
```

```python
for container in containers:
    job_titles = container.find_all("h2", {"class": "css-m604qf"})
    job_title = job_titles[0].text.strip()
    
    comp_names = container.find_all("a", {"class": "css-17s97q8"})
    comp_name = comp_names[0].text.strip()
    
    job_types = container.find_all("a", {"class": "css-n2jc4m"})
    job_type = job_types[0] .text.strip()
    
#     print(job_title)
#     print(comp_name)
#     print(job_type)
#     print()
    file1.write(job_title + "," + comp_name + "," + job_type + "\n")
file1.close()
```
